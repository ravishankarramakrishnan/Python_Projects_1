```python
# Importing the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import the dataset
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
movies = df
```

```python
movies = movies.join(movies.genres.str.get_dummies('|'))
```

```python
movies
```

```python
# Identifying the Similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(movies.iloc[0:100,3:])
```

```python
cos_sim[0]
```

```python
# Top Similar Films
sim_top5 = np.argsort(cos_sim[19])[-5:][::-1]
```
